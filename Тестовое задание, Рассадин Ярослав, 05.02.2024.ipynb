```python
import pandas as pd

df = pd.read_excel('Тестовое задание, Рассадин Ярослав.xlsx', usecols=['Тема письма ', 'Отправлено', 'Доставлено', 'Открытия', 'Клики', 'Отписки', 'День недели'])

df.rename(columns={'Тема письма ': 'Тема письма'}, inplace=True)

df['Delivery Rate'] = df['Доставлено'] / df['Отправлено'] * 100
df['Open Rate'] = df['Открытия'] / df['Доставлено'] * 100
df['Click to Open Rate'] = df['Клики'] / df['Открытия'] * 100
df['Unsubscribe Rate'] = df['Отписки'] / df['Доставлено'] * 100

df.info()
df.head()
```

```python
# Вычисление метрик
delivery_rate = df['Delivery Rate'].sum()
open_rate = df['Open Rate'].sum()
click_to_open_rate = df['Click to Open Rate'].sum()
unsubscribe_rate = df['Unsubscribe Rate'].sum()

# Определение лучшей темы
best_topic = df.groupby('Тема письма')['Открытия'].sum().idxmax()

# Группировка данных по дню недели и вычисление средних значений метрик
mean_metrics_by_weekday = df.groupby('День недели').mean()

# Определение лучшего дня недели
best_day = mean_metrics_by_weekday[['Отправлено', 'Открытия', 'Клики']].sum(axis=1).idxmax()


# Вывод результатов
print('Delivery rate:', delivery_rate)
print('Open rate:', open_rate)
print('Click to Open rate:', click_to_open_rate)
print('Unsubscribe rate:', unsubscribe_rate)
print('Лучшая тема:', best_topic)
print('Лучший день недели:', best_day)
```
